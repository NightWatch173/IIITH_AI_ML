```
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

def dataCleaningWords(wordList):
    
    tmpList = []
    # 1. Removal of Punctuation Marks 
    tmpList = [each for each in wordList if each not in string.punctuation]
    
    # 2. Removal of Stopwords
    tmpList2 = []
    tmpList2 = [each for each in tmpList if each not in stopwords.words('english')]

    # 3. Lemmatisation 
    tmpList3 = []
    tmpList3 = [WordNetLemmatizer().lemmatize(each, pos="v") for each in tmpList2]
    return tmpList3
  
  
def dataCleaningSent(sentList):
  cleanedSentList = []
  for eachSent in sentList:
    cleanedSentList.append(' '.join(dataCleaningWords(word_tokenize(eachSent))))
  return cleanedSentList
```

### a) Identify 10 handcrafted features of the kind described by the  story above- 2 Marks
 




```
# UniqueWords    AvgSentLenth  mcwFeature   Label 

from nltk.probability import FreqDist
finalVectorCombined = []

def buildFinalVectorCombined(storyList, label):
  for eachStory in storyList:
    sentLenSum = 0
    wordsList = []
    for eachSent in eachStory:
      wordsList.extend(word_tokenize(eachSent))
      sentLenSum += len(word_tokenize(eachSent))

    avgSentLength = sentLenSum/len(eachStory)
    uniqueWordsLength = len(set(wordsList))

    fdist = FreqDist(wordsList)
    mcw = sum([y for x,y in fdist.most_common(50)])
    mcwFeature = 0
    if(len(wordsList)!=0):
      mcwFeature = mcw/len(wordsList)

    # print("avgSentLength =", avgSentLength, "uniqueWordsLength = ", uniqueWordsLength, "mcwFeature = ", mcwFeature)
    finalVectorCombined.append([uniqueWordsLength,avgSentLength, mcwFeature, label])
    
buildFinalVectorCombined(antonStories, 0)
buildFinalVectorCombined(guyStories, 1)
```
