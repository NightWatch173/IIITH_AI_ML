As a data scientist, you will use sets of data in the form of dictionaries, DataFrames, or any other data type. When working with those, you might want to save them to a file, so you can use them later on or send them to someone else. This is what Python's pickle module is for: it serializes objects so they can be saved to a file, and loaded in a program again later on. Further, you can even pickle your machine learning model. However do be aware that, pickle is illadvised, when saving in one particular programming language and loading on another. The same would be a concern even in case of saving and loading across different versions of Python. The idea is that this character stream contains all the information necessary to reconstruct the object in another python script.









### 1. Saving a dictionary into pickle file

```
#@title
# Importing the Library
import pickle
```

Creating a dictionary with state and it's capitals.

Keys are states and values are capitals

```
Capitals = {"Assam":"Dispur", "Gujarat":"Gandhinagar", "Himachal Pradesh":"Shimla"}
```

```
Capitals
```

Dumping Capitals dictionary into a pickle file

```
#@title
pickle.dump(Capitals, open( "capitals.pkl", "wb" ) )
```

### 2. Load the dictionary back from the pickle file

```
Capitals = pickle.load( open( "capitals.pkl", "rb" ) )
```
