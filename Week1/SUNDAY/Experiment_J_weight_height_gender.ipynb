#### Importing Required Packages

```
Dataset = "weights_heights_gender.csv"
```

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
def gender_converter(s):
  s = s.lower()
  if s == "male":
    return 1
  elif s == "female":
    return 2
  return -1
```

```
data = pd.read_csv(Dataset,converters={"Gender":gender_converter})
print(data.size)
```

```
data.head()
```

```
data.tail()
```

#### Loading the data

```
feature_names = ['Height','Weight']
X = data[feature_names]
y = data['Gender']
```

### Splitting the data into train and test sets 

```
from sklearn.model_selection import train_test_split
```

```
 # Training and testing set ratio is 70 : 30
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```
# Let us see the size of train and  test sets
X_train.shape, X_test.shape
```

```
# Let us see first five rows of the training data

X_train[:5]
```

### Training a  Knn Classifier 

```
from sklearn.neighbors import KNeighborsClassifier
```

```
neigh = KNeighborsClassifier(n_neighbors=3)
```

```
# Training or fitting the model with the train data
neigh.fit(X_train,y_train)
```

```
# Testing the trained model
neigh.predict(X_test)
```

```
# Calculating the score
neigh.score(X_test,y_test)
```
